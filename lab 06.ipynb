```python
import mlrose
import numpy as np

# Define the fitness function for the N-Queens problem
def queens_max(position):
    no_attack_on = 0
    queen_not_attacking = 0
    n = len(position)

    for i in range(n):
        for j in range(i + 1, n):
            # Check if queens are not in the same row or diagonal
            if position[i] != position[j] and abs(position[i] - position[j]) != abs(i - j):
                no_attack_on += 1

    queen_not_attacking = no_attack_on
    return queen_not_attacking

# Define the problem
objective = mlrose.CustomFitness(queens_max)
problem = mlrose.DiscreteOpt(length=8, fitness_fn=objective, maximize=True, max_val=8)

# Define initial state and simulated annealing parameters
initial_position = np.array([4, 0, 1, 5, 2, 6, 3, 7])
schedule = mlrose.ExpDecay()
max_attempts = 500
max_iters = 5000

# Run simulated annealing
best_position, best_objective = mlrose.simulated_annealing(
    problem=problem,
    schedule=schedule,
    max_attempts=max_attempts,
    max_iters=max_iters,
    init_state=initial_position
)

# Print the results
print("The best position found is:", best_position)
print("The number of queens that are not attacking each other is:", best_objective)
```

```python
import numpy as np

def calculate_attacks(position):
    attacks = 0
    n = len(position)

    for i in range(n):
        for j in range(i + 1, n):
            if position[i] == position[j] or abs(position[i] - position[j]) == j - i:
                attacks += 1
    return attacks

def simulated_annealing(n, max_iters=1000, initial_temp=100, cooling_rate=0.99):
    # Generate initial position randomly
    current_position = np.random.permutation(n)
    current_attacks = calculate_attacks(current_position)
    temperature = initial_temp

    for i in range(max_iters):
        # If there's no attack, solution is found
        if current_attacks == 0:
            break

        # Generate a new candidate by swapping two queens
        new_position = current_position.copy()
        i, j = np.random.choice(n, 2, replace=False)
        new_position[i], new_position[j] = new_position[j], new_position[i]

        # Calculate the number of attacks in the new position
        new_attacks = calculate_attacks(new_position)

        # Decide whether to accept the new position
        if new_attacks < current_attacks or np.random.rand() < np.exp((current_attacks - new_attacks) / temperature):
            current_position = new_position
            current_attacks = new_attacks

        # Decrease the temperature
        temperature *= cooling_rate

    return current_position, current_attacks

n = 8  # Number of queens (for the 8-Queens problem)
solution, attacks = simulated_annealing(n)

print("Best position found:", solution)
print("Number of attacks:", attacks)
```
